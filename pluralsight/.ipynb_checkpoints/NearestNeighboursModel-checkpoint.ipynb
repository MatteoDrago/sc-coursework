```python
import pandas as pd
dataFile='/Users/tommy/Downloads/BX-CSV-Dump/BX-Book-Ratings.csv'
data=pd.read_csv(dataFile,encoding ="ISO-8859-1",sep=";",header=0,names=["users", "isbn", "ratings"])
```

```python
data.head()
```

```python
bookFile='/Users/tommy/Downloads/BX-CSV-Dump/BX-Books.csv'
books=pd.read_csv(bookFile,encoding="ISO-8859-1",sep=";",header=0,error_bad_lines=False,usecols=[0,1,2],index_col=0,names=["isbn","title","author"])
```

```python
books.head()
```

```python
def bookMeta(isbn):
    title = books.at[isbn,"title"]
    author = books.at[isbn,"author"]
    return title,author
bookMeta("0671027360")
```

```python
def faveBooks(user,N):
    userRatings = data[data["users"]==user]
    sortedRatings = pd.DataFrame.sort_values(userRatings,by="ratings",ascending=False)[:N]
    sortedRatings["title"] = sortedRatings["isbn"].apply(bookMeta)
    return sortedRatings
```

```python
data = data[data["isbn"].isin(books.index)]
```

```python
faveBooks(204622,5)
```

```python
data.shape
```

```python
usersPerISBN = data.isbn.value_counts()
```

```python
usersPerISBN.head(10)
```

```python
ISBNsPerUser = data.users.value_counts()
```

```python
data = data[data["isbn"].isin(usersPerISBN[usersPerISBN>10].index)]
```

```python
data = data[data["users"].isin(ISBNsPerUser[ISBNsPerUser>10].index)]
```

```python
userItemRatingMatrix=pd.pivot_table(data,values="ratings",index=['users'],columns=["isbn"])
```

```python
userItemRatingMatrix.head()
```

```python
userItemRatingMatrix.head(50)
```

```python
user1 = 204622
user2 = 255489
```

```python
user1Ratings = userItemRatingMatrix.transpose()[user1]
user1Ratings.head()
```

```python
user2Ratings = userItemRatingMatrix.transpose()[user2]   
```

```python
from scipy.spatial.distance import hamming
hamming(user1Ratings, user2Ratings)
```

```python
import numpy as np
def distance(user1,user2):
    try:
        user1Ratings = userItemRatingMatrix.transpose()[user1]
        user2Ratings = userItemRatingMatrix.transpose()[user2]
        distance = hamming(user1Ratings,user2Ratings)
    except:
        distance = np.NaN
    return distance
```

```python
user = 204622
allUsers = pd.DataFrame(userItemRatingMatrix.index)
allUsers = allUsers[allUsers.users!=user]
```

```python
allUsers["distance"] = allUsers["users"].apply(lambda x: distance(user,x))
```

```python
allUsers.head()
```

```python
K = 10
KnearestUsers = allUsers.sort_values(["distance"],ascending=True)["users"][:K]
```

```python
KnearestUsers
```

```python
def nearestNeighbours(user,K=10):
    allUsers = pd.DataFrame(userItemRatingMatrix.index)
    allUsers = allUsers[allUsers.users!=user]
    allUsers["distance"] = allUsers["users"].apply(lambda x: distance(user,x))
    KnearestUsers = allUsers.sort_values(["distance"],ascending=True)["users"][:K]
    return KnearestUsers
    
```

```python
KnearestUsers = nearestNeighbours(user)
```

```python
NNRatings = userItemRatingMatrix[userItemRatingMatrix.index.isin(KnearestUsers)]
```

```python
NNRatings
```

```python
avgRating = NNRatings.apply(np.nanmean).dropna()
```
